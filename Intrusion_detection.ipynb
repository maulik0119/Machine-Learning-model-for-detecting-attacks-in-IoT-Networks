# **Data Normalization**


```python
# Numeric attributes
num_col = list(df.select_dtypes(include='number').columns)
num_col.remove('id')
num_col.remove('label')
print(num_col)
```

```python
# Minmax scaler for normalizing data
minmax_scale = MinMaxScaler(feature_range=(0, 1))
def normalization(dff,col):
  for i in col:
    arr = dff[i]
    arr = np.array(arr)
    dff[i] = minmax_scale.fit_transform(arr.reshape(len(arr),1))
  return dff
```

```python
# data before normalization
df.head()
```

```python
df = normalization(df.copy(),num_col)
```

```python
# data after normalization
df.head()
```

# **Label Encoding**

## **Binary Labels**

```python
# changing attack labels into two categories 'normal' and 'abnormal'
binary_label = pd.DataFrame(df.label.map(lambda x:'normal' if x==0 else 'abnormal'))
```

# **BINARY CLASSIFICATION**

## **Data Splitting**



```python
X = binary_df.drop(columns=['label'],axis=1)
Y = binary_df['label']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.20, random_state=50)
```

## **Linear Regression**

```python
lr_bin = LinearRegression(normalize=False)
lr_bin.fit(X_train, y_train)
```

```python
y_pred = lr_bin.predict(X_test)
```

```python
round = lambda x:1 if x>0.6 else 0
vfunc = np.vectorize(round)
y_pred = vfunc(y_pred)
```

```python
print("Mean Absolute Error - " , metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error - " , metrics.mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error - " , np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("R2 Score - " , metrics.explained_variance_score(y_test, y_pred)*100)
print("Accuracy - ",accuracy_score(y_test,y_pred)*100)
```

```python
cls_report= classification_report(y_true=y_test, y_pred=y_pred,target_names=le1.classes_)
```

```python
print(cls_report)
```
